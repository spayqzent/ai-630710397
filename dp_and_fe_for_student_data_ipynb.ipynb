```python
# เติมค่าหายในคอลัมน์ 'อายุ' ด้วยค่าเฉลี่ย
mean_age = df['อายุ'].mean()
df['อายุ'] = df['อายุ'].fillna(mean_age)

# เติมค่าหายในคอลัมน์ 'GPA' ด้วยค่าเฉลี่ย
mean_gpa = df['GPA'].mean()
df['GPA'] = df['GPA'].fillna(mean_gpa)
```

```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
```

Q1 คือค่าที่แสดงตำแหน่ง 25% ของข้อมูล

Q3 คือค่าที่แสดงตำแหน่ง 75% ของข้อมูล

การใช้ IQR จะช่วยตรวจจับค่า outlier ได้ โดยการกำหนดค่าที่อยู่นอกช่วง Q1 - 1.5 * IQR ถึง Q3 + 1.5 * IQR ว่าเป็น outliers

```python
Q1,Q3
```

```python
IQR = Q3 - Q1
```

```python
IQR
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
```

```python
# เติมค่า NaN ในคอลัมน์ 'เพศ' ด้วยค่าที่พบมากที่สุด (mode)
mode_gender = df['เพศ'].mode()[0]
df['เพศ'] = df['เพศ'].fillna(mode_gender)
df['เพศ'] = df['เพศ'].replace({'M': 'ชาย', 'F': 'หญิง', 'ช': 'ชาย', 'ญ': 'หญิง'})
df['เพศ'] = df['เพศ'].replace({'Male': 'ชาย', 'Female': 'หญิง'})

```

แทนที่ค่าผิดปกติ เช่น 'M' → 'ชาย', 'F' → 'หญิง', 'ช' → 'ชาย', 'ญ' → 'หญิง' เพื่อให้ข้อมูลมีรูปแบบที่ถูกต้อง

แทนที่ค่าภาษาอังกฤษ เช่น 'Male' → 'ชาย' และ 'Female' → 'หญิง' เพื่อให้ข้อมูลทั้งหมดอยู่ในรูปแบบเดียวกัน
